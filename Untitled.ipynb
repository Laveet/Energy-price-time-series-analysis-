```python
import pandas as pd
import numpy as np
df = pd.read_csv('energy_dataset.csv')



df['time'] = pd.to_datetime(df['time'], utc=True)  # parse with timezone


df['time'] = df['time'].dt.tz_localize(None)

df.set_index('time', inplace=True)
```

```python
df = df.dropna(axis=1, how='all')
df = df.loc[:, ~((df.fillna(0) == 0).all())]
```

```python
import matplotlib.pyplot as plt

def plot_columns_interactive(df):
    while True:
        print("\nAvailable columns:")
        print(list(df.columns))  # just for column names
        
        col = input("Enter column name to plot (or type 'quit' to exit): ").strip()
        
        if col.lower() in ['end']:
            print("Exiting...")
            break
        
        if col in df.columns:
            df[col].plot(figsize=(12, 4), title=f"{col}")
            plt.xlabel("Time")
            plt.ylabel(col)
            plt.grid(True)
            plt.show()
        else:
            print(f"Column '{col}' not found. Please try again.")
```

```python
plot_columns_interactive(df)
```

```python
df = df[df.index.year != 2014]
df['year']= df.index.year
df['month']= df.index.month
```

```python
monthly_load = df.groupby(['year','month'])['total load actual'].sum().unstack(level=0)
```

```python
monthly_load
```

```python
for year in monthly_load.columns:
    plt.plot(monthly_load.index, monthly_load[year], marker='o', label=str(year))

plt.xlabel('Month')
plt.ylabel('Total Load Actual')
plt.title('Monthly Total Load Actual by Year')
plt.xticks(ticks=range(1,13), labels=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
plt.legend(title='Year')
plt.grid(True)
plt.show()
```

```python
monthly_load.plot(kind='bar', figsize=(14,6))

plt.title("Monthly Total Load per Year")
plt.xlabel("Month")
plt.ylabel("Total Load Actual")
plt.xticks(ticks=range(12), labels=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'], rotation=0)
plt.legend(title='Year')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Ensure datetime index and frequency is set
df_Hourly= df.asfreq('H')  # Assuming hourly data

# Decompose
decomposition = seasonal_decompose(df['price actual'], model='additive')

# Plot the components
decomposition.plot()
plt.suptitle("Decomposition of Total Load Actual", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
window_size= 24*7

rolling_mean= df['total load actual'].rolling(window=window_size).mean()
rolling_std = df['total load actual'].rolling(window=window_size).std()

plt.figure(figsize=(14, 5))
plt.plot(df['total load actual'], color='lightgray', label='Original')
plt.plot(rolling_mean, color='blue', label='Rolling Mean (7 days)')
plt.plot(rolling_std, color='red', label='Rolling Std (Volatility)')
plt.legend()
plt.title("Rolling Mean and Std of Total Load (7-day window)")
plt.show()
```
